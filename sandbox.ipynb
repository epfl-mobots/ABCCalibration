# Nb used to explore the power distribution

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', 2000)
```

## Configuration

```python
path = 'max_pwm_powers.csv'
abcs = ['abc02','abc03','abc04','abc23']
```

## Main code

```python
df = pd.read_csv(path)
df = df[df['board_id'].isin(abcs)]
df.drop(columns=['base'], inplace=True)
print(df)

all_powers = np.array([])
for idx, row in df.iterrows():
    row.drop(labels=['board_id'], inplace=True)
    all_powers = np.concatenate((all_powers, row.to_numpy()))

avg_power = round(all_powers.mean(), 3)
std_power = all_powers.std()

print(f"Average power: {avg_power}, std of power: {std_power}")
```
